## Concepts

1. What is a SQL?
    - R.: SQL is a Standardize Language for interacting with R.D.B.M.S. (Relational Databases Management System).
    - Used to perform C.R.U.D. (Create, Read, Update, Delete) operations, as well as other administrative taks (user management, security, backup, etc).
    - Used to define tables and schemas structures.
    - SQL code used on one R.D.B.M.S is not always portable to another without modification.

2. What is a database?
    - Any collection of related information. There are two types of Databases,
        **relational databases** and **non-relational databases**. there are special that helps users create and maintain a databases called **Databases Management System** (D.B.M.S). DBMS allows you to perform the C.R.U.D. operations and other administrative tasks.
        
3. What are relational databases?
    - Organize data into one or more tables.
    - Each table has columns and rows.
    - A unique Key identifies each row.
4. What is a query?
    - A query is a set of instructions given to the R.D.B.M.S. (write in SQL) that tell the R.D.B.M.S. what information you want it to retrive for you, **the goal is to only get the data you need**.
    - Queries are requests made to the database management system for specific information.
    - As the database structure becomes more complex, getting the specific information we want becomes more difficcult.
    
5. What are the **Primary Key** and **Foreing Key**?
    - **Primary Key**: an unique attribute defines the row in the database.
    - **Foreign Key** an attribute that we can store on a database table that will link us to another database, it's another **Primary Key** inside anoter table.
    - **Composite Key**: Key that needs two attibutes.
    
   

## Another concepts

1. SQL commands
    - DDL: Data Definition Language, used for defining database schemas, for instance, CREATE, DROP, ALTER, TRUNCATE,...
    - DML: Data Manipulation Language, used for inserting, updating, and deleting data from databases (INSERT, UPDATE, DELETE, MERGE, ...).
    - DCL: Data Control Language, used for controlling access to the data in the databases, and user and permissions management (GRANT, REVOKE, ...).
    - CTL: Control Transaction Language (COMMIT, ROLLBACK, SAVEPOINT, ...).
    - DQL: Data Query Languege, used to query the data for information and get information that is already there (SELECT, ...).
    - Parent child ralationship: a parent-child relationship between two tables can be created only when there is a **PRIMARY KEY**  in one table and **FOREIGN KEY** in another table. (w33resource)
    - Child table: a child table inherits the **PRIMARY KEY** colums of its parent table. (oracle)
    - Parent table: the table with the **FOREING KEY** is called the child table, and the table with the **PRIMARY KEY** is called the referenced or parent table.(w3schools)

## Wildcard

A wildcard character is used to substitute one or more characters in a string. Wildcard characters are used with the **LIKE** operator. The **LIKE** operator is used in a **WHERE** clause to search or specified pattern in a columnn. (w3schools)

- '%' -> represents zero or more characters.
- '_' -> reprensents a single character.

**Syntax**
1.\

**FROM** table_name\

When you know that unknown characters are on the left side.

2.\
**SELECT** *\

When you know that unknown characters are on the right side.

## JOIN

1. What is a Join?
    - It a combine a row from two or more tables dases in related columns, is a command clause that combines records from two or more tables in a database (Coursera).
    - **INNER JOIN**: combine records from two table whenever these are matching values in a field common to both tables (Microsoft Support)
    - **OUTERJOIN**: joins that return **matched** and **unmatched** values from either or both tables.
    - **LEFT JOIN**: return all records from left table and **matched** records from the **right tables**.
    - **RIGHT JOIN**: this join returns **all the rows** of the table on the right side of the join and **matching row** for the table on the left side, the result-set will contain **NULL** (GeeksforGeeks).
    - **FULL OUTERJOIN**: return all records when there is a match in left or right table records.


1.\
**SELECT** table_1.attribute_1, table_1.attribute_2\
**FROM** table_1\
**JOIN** table_2\
**ON** table_1.attribute_n(**PRIMARY KEY**) = table_2.attribute_n(**FOREIGN KEY**);

- table_1.attribute_n and table_2.attribute_n it's the same attribute but in differents tables.

2.\
**SELECT** table_1.attribute_1, table_1.attribute_2\
**FROM** table_1\
**LEFT JOIN** table_2\
**ON** table_1.attribute_n(**PRIMARY KEY**) = table_2.attribute_n(**FOREIGN KEY**);

3.\

**FROM** table_1\
**RIGHT JOIN** table_2\
**ON** table_1.attribute_n(**PRIMARY KEY**) = table_2.attribute_n(**FOREIGN KEY**);
